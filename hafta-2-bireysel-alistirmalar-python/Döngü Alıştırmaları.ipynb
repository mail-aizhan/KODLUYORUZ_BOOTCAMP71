<a href="https://www.wikizero.com/en/Compartmental_models_in_epidemiology">SIR (susceptible-infected-recovered) modeli</a>, salg캼n hastal캼klar캼n yay캼lmas캼n캼 incelemekte kullan캼lan modellerin en sadelerindendir. $$\begin{array}{rcl}S_{t+1} &=& S_t - a I_t S_t \\ I_{t+1} &=& I_t + a I_t S_t - b I_t \\ R_{t+1} &=& R_t + bI_t \end{array}$$

Burada $a$ parametresi, kullan캼lan zaman birimi i칞inde (g칲n, hafta, ay) hastal캼kla temas s캼kl캼캼 veya bula릀a olas캼l캼캼, $b$ parametresi ise ortalama iyile릀e oran캼d캼r. Buna g칬re 1/$b$ de른ri hastal캼캼n iyile릀esi i칞in bir zaman 칬l칞e를 sa륿ar, bu bilgiyle de $b$ i칞in bir tahmin yap캼labilir.

SIR modelini kullanarak her zaman ad캼m캼nda S, I, ve R kategorisindekilerin oran캼n캼 listeleyen bir program yaz캼n. Veriler hasta olanlar캼n oran캼 %0.1'in 칲zerinde oldu릇 s칲rece, ama en fazla 100 zaman ad캼m캼 g칬sterecek 른kilde listelensin. Her zaman ad캼m캼nda $S_t + I_t + R_t$ toplam캼n캼 da ekrana bast캼r캼n. Bu toplam캼n sabit kalmamas캼 program캼n캼zda bir hata oldu릇nu g칬stergesidir.

# Babil Y칬ntemiyle Karek칬k Bulma

```python
import math

def b_karekok(n):
    value2 = float(input("캻lk tahmin: "))
    tol = 1*10**-10
    value1 = math.sqrt(n)
    counter = 0
    while abs(value2-value1)>=tol:
        value2 = (1/2)*(value2+(n/value2))
        counter += 1
    print(f"Karek칬k ~ {value2}")
    print(f"{counter} iterasyon")


b_karekok(float(input("Karek칬k칲 al캼nacak say캼: ")))
```
